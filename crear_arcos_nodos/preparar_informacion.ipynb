```python
import pandas as pd

```

```python
bd_inversionistas = pd.read_excel('Base-inicial.xlsx')
bd_inversionistas.head(5)
```

```python
import numpy as np
bd_inversionistas = bd_inversionistas[~bd_inversionistas["Principales inversionistas"].isna()]
nodos = bd_inversionistas["Organization Name"][~bd_inversionistas["Principales inversionistas"].isna()].unique()
relaciones = {}
df_inversionistas = {SOURCE: TARGET.split(",") if str(TARGET) != 'nan' else np.nan for SOURCE, TARGET in zip (bd_inversionistas["Organization Name"], bd_inversionistas["Principales inversionistas"]) }
df_inversionistas

lista_tuplas = []
for empresa in df_inversionistas:
    if str(df_inversionistas[empresa]) == 'nan': # relaciÃ³n consigo misma
        None
    else:
        for target in df_inversionistas[empresa]:
            lista_tuplas.append((empresa,target.strip()))

```

```python
bd_inversionistas_total = pd.read_excel('Base-inicial.xlsx')
arcos = pd.DataFrame(lista_tuplas, columns=['SOURCE','TARGET'])
total_nodos = set(list(arcos.SOURCE.unique())+list(arcos.TARGET.unique()))
nodos_final = pd.DataFrame(total_nodos).rename(columns={0:"Organization Name"}).merge(bd_inversionistas_total, how="left", on="Organization Name").reset_index().rename(columns={'index':"Id","Organization Name":"Label"})
nodos_final
```

```python
arcos_final = arcos.rename(columns={"SOURCE":"LabelSource","TARGET":"LabelTarget"}).merge(
                nodos_final[["Id","Label"]].rename(columns={"Id":"SOURCE"}), left_on="LabelSource",right_on="Label").merge(
                    nodos_final[["Id","Label"]].rename(columns={"Id":"TARGET"}), left_on="LabelTarget",right_on="Label").reset_index()

arcos_final = arcos_final.rename(columns={"index":"Id"})[["Id","SOURCE","TARGET","LabelSource","LabelTarget"]]
arcos_final
```

```python
# Exportar la informaciÃ³n de arcos y nodos
nodos_final.to_csv("nodos.csv", index=False)
arcos_final.to_csv("arcos.csv", index=False)
```
